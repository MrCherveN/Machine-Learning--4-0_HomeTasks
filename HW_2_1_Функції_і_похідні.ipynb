### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return 10*x - 5

def diff_func(f, x, h):
  return (f(x + h) - f(x)) / h

x = np.linspace(0, 10, 100)

f_lin = f(x)
#derivative_f_lin = [10]*len(x) # альтернативний варіант вручну
derivative_f_lin = diff_func(f, x, len(x))

plt.figure(figsize=(10, 5))
plt.plot(x, f_lin, label='f(x) = 10x-5', color='blue')
plt.plot(x, derivative_f_lin, label="f'(x) = 10", color='green')
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show();
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp

x = sp.symbols('x')

y1 = sp.sin(x)
dy1 = sp.diff(y1, x)

print("Функція:", y1)
print("Похідна:", dy1)

y2 = sp.ln(x)
dy2 = sp.diff(y2, x)

print("Функція:", y2)
print("Похідна:", dy2)

f1 = sp.lambdify(x, y1, 'numpy')
df1 = sp.lambdify(x, dy1, 'numpy')
f2 = sp.lambdify(x, y2, 'numpy')
df2 = sp.lambdify(x, dy2, 'numpy')

xs = np.linspace(0.1, 20, 100)

plt.figure(figsize=(10, 5))
plt.plot(xs, f1(xs), label=f'Функція: {y1}', color='blue')
plt.plot(xs, df1(xs), label=f'Похідна функції: {y1}', color='red')

plt.plot(xs, f2(xs), label=f'Функція: {y2}', color='green')
plt.plot(xs, df2(xs), label=f'Похідна функції: {y2}', color='yellow')

plt.title("Графік функцій та їх похідних")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show();
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

y = x**2 - 5*x + 1
dy = sp.diff(y, x)

print("Функція:", y)
print("Похідна:", dy)

f = sp.lambdify(x, y, 'numpy')
df = sp.lambdify(x, dy, 'numpy')

xs = np.linspace(0, 100, 1000)

plt.figure(figsize=(10, 5))
plt.plot(xs, f(xs), label=f'Функція: {y}', color='blue')
plt.plot(xs, df(xs), label=f'Похідна: {dy}', color='red')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show();
```
