**1. Гістограма розподілу рахунків**
Завантажте датасет Tips (Чайові) з файла у матеріалах `tips.csv`.
Створіть гістограму для колонки `total_bill` з 20 bins.
Опціонально можна додати підписи осей ОХ, ОУ, а також назву графіку.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv('/content/tips.csv')
df.head()
```

```python
df['total_bill'].hist(bins=20, edgecolor='black', figsize=(10, 6))
plt.title('Розподіл загальної суми рахунку')
plt.xlabel('Загальна сума рахунку (долари)')
plt.ylabel('Кількість')
plt.show();
```

---

**2. Box plot чайових за днями тижня**

Створіть box plot, який показує розподіл чайових (`tip`) за днями тижня (`day`).
В який день навище медіанне (середнє) значення продажів? Зелена горизонтальна лінія всередині кожного box-plot.


```python
df.boxplot(column='tip', by='day', figsize=(10, 6))
plt.title('Розподіл чайових за днями тижня')
plt.xlabel('День')
plt.ylabel('Чайові')
plt.suptitle('')
plt.show();
```

Відповідь: в неділю спостерігається найвище медіанне (середнє) значення продажів.

---

**3. Scatter plot: рахунок vs чайові**

Створіть діаграму розсіювання для `total_bill` та `tip`.

```python
df.plot.scatter(x='total_bill', y='tip', grid=True, figsize=(10, 6))
plt.title('Залежність між загальною сумою рахунку та чайовими')
plt.xlabel('Загальна сума рахунку')
plt.ylabel('Чайові')
plt.show()
```

---

**4. Bar chart: середній рахунок за часом**

Створіть стовпчасту діаграму середнього рахунку за часом дня (`time`).

```python
df.groupby('time')['total_bill'].mean().plot(kind='bar', edgecolor='black', figsize=(10, 6))
plt.title('Середній рахунок за часом дня')
plt.xlabel('Час дня')
plt.ylabel('Середній рахунок')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show();
```

---

**5. Інтерактивна діаграма розсіювання**

Створіть з допомогою бібліотек Plotly інтерактивний scatter plot для `total_bill` vs `tip` з кольором за статтю (`sex`).


```python
fig = px.scatter(df, x='total_bill', y='tip', color='sex', title='Загальний рахунок vs Чайові',
                 labels={'total_bill': 'Загальний рахунок', 'tip': 'Чайові', 'sex': 'Cтать'},
                 width=900,
                 height=600
                 )
fig.show();
```

---

**6. Інтерактивний box plot: чайові за днями та статтю**

Створіть інтерактивний box plot для чайових за днями тижня з розбивкою за кольором по статі.

```python
fig = px.box(df, x='day', y='tip', color='sex',title='Чайові за днями та статтю', labels={'day': 'День тижня', 'tip': 'Чайові', 'sex': 'Cтать'},  width=900,
                 height=600)
fig.show();
```
